```python
numeric_data = df.select_dtypes(include=['float64', 'int64'])
correlation = numeric_data.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
nationality_counts = df['Nationality'].value_counts()
nationality_counts.plot(kind='bar', figsize=(12, 8))
plt.title("Number of Songs by Nationality")
plt.xlabel("Nationality")
plt.ylabel("Count")
plt.show()
```

```python
# Select relevant columns
selected_features = ['Energy', 'Loudness', 'Danceability', 'Instrumentalness', 'Rank']

#scatter plots for selected audio features
sns.set(style="whitegrid")
sns.pairplot(df[selected_features], diag_kind='kde')
plt.suptitle("Scatter Plots of Audio Features vs Rank", y=1.02)
plt.show()
```

```python
artist_points = df.groupby('Artists')['Points (Total)'].sum().sort_values(ascending=False)


# Choose The Top N Artists (e.g., top 10)
top_n = 10
top_artists = artist_points.head(top_n)

# Get the titles of popular songs for the top artists
popular_songs = df[df['Artists'].isin(top_artists.index)][['Artists', 'Title']]

# Create a bar plot
plt.figure(figsize=(8, 6))
plt.barh(top_artists.index, top_artists.values, color='purple')
plt.xlabel('Total Points')
plt.ylabel('Artists')
plt.title('Top Artists and Their Popular Songs')
plt.gca().invert_yaxis() 
plt.show()
```

```python
energy_danceability = df[['Energy', 'Danceability']]

# Create a scatter plot
plt.figure(figsize=(8, 8))
plt.scatter(energy_danceability['Energy'], energy_danceability['Danceability'], color='tab:red', alpha=0.5)
plt.xlabel('Energy')
plt.ylabel('Danceability')
plt.title('Energy vs. Danceability of Songs')
plt.grid(True)
plt.show()
```

```python
df = df.drop_duplicates(subset=['Title', 'Points (Total)'], keep='first')

# Select columns for analysis
top_songs = df[['Title', 'Points (Total)']].sort_values('Points (Total)', ascending=False).head(20)

# Create Bar Plot with adjusted plot size
plt.figure(figsize=(8, 4))  
sns.barplot(data =top_songs, x='Points (Total)', y='Title')
plt.title('Top Songs by Points')
plt.xlabel('Points (Total)')
plt.ylabel('Title')
plt.show()
```

```python
# 4. Trend Over Time: Total Points
daily_points = df.groupby('Date')['Points (Total)'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(data=daily_points, x='Date', y='Points (Total)', marker='o', color='skyblue')
plt.title('Trend of Total Points Over Time')
plt.xlabel('Date')
plt.ylabel('Total Points')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top_nationalities = df['Nationality'].value_counts().head(10)
sns.barplot(x=top_nationalities.index, y=top_nationalities.values, palette='muted')
plt.title('Top 10 Nationalities in Dataset')
plt.ylabel('Count')
plt.xlabel('Nationality')
plt.xticks(rotation=45)
plt.show()
```
